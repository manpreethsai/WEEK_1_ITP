### Problem 1:

1. Write a program to calculate area of shapes. Your program should be capable of calculating the area of a square, rectangle, triangle and a circle. The user should be presented with options to select the shape. Based on which shape is chosed by the user, the program should ask for the appropriate input and print the resulting area on the screen.

When the program is run, the screen should display something like this:

**Which shape would you like to calculate the area for? Please enter the option number-**<br>
**1. Square**<br>
**2. Rectangle**<br>
**3. Triangle**<br>
**4. Circle**<br>
**Enter Option: _**

Say the user enters the option 1.

**Please enter the length of a side: _**

If the user enters a value of 5. The output should be:

**The area of the square is 25**

This program should indicate that the input is invalid if the user enters a character instead of a number as input. For instance if the user enters a value of ‘a’ instead of 5 in the previous example the program should prompt:

**Invalid input, please enter a number: _**


```python
def input_validity(input_value1, input_value2='', input_value3=''):
    if (input_value1 in '1234567890' and input_value2 in '1234567890'and input_value3 in '1234567890'):
        
        return True
    
    else:
     
        return False


def square_area(side):
    
        area_square = side**2
        print('THE AREA OF THE SQUARE IS: ', area_square)
        return
    
def rectangle_area(length, breadth):
    
        area_rectangle = length*breadth
        print('THE AREA OF THE RECTANGLE IS: ', area_rectangle)
        return
    
def triangle_area(base, height):
    
        area_triangle = 0.5*base*height
        print('THE AREA OF THE TRIANGLE IS: ', area_triangle)
        return
    
def circle_area(radius):
    
        area_circle = (22/7)*(r**2)
        print('THE AREA OF THE CIRCLE IS: ', area_circle)
        return
    
print('Which shape would you like to calculate the area for?' + '\n' + 'Please enter the option number-' + '\n' + '1. Square' + '\n' + '2. Rectangle' + '\n' + '3. Triangle' + '\n' + '4. Circle \n')

choice = str(input('ENTER THE SHAPE OPTION: \n'))

while(True):
    if (choice != '1' and  choice !='2' and choice !='3' and choice !='4'):
        #print('PLEASE ENTER A VALID SHAPE OPTION')
        choice = input('PLEASE RE-ENTER A VALID SHAPE OPTION: \n')
        continue
   
############################################################
    elif choice=='1':
        print('Square area is being calculated.')
        side = input('Enter the side of the square: \n')
        check1 = input_validity(input_value1 = side)
        if check1==True:
            square_area(float(side))
            break
        
        else:
            while(True):
                print('INVALID INPUT. PLEASE ENTER A POSITIVE NUMBER\n')
                new_side = input('Enter the side of the square: \n')
                check1_repeat = input_validity(input_value1 = new_side)
                if (check1_repeat==True):
                    square_area(float(new_side))
                    break
        break
############################################################                
    elif choice=='2':
        print('Rectangle area is being calculated.')
        length = input('Enter the length of the rectangle: \n')
        breadth = input('Enter the breadth of the rectangle: \n')
        check1 = input_validity(input_value1 = length, input_value2 = breadth)
        #check2 = input_validity(breadth)
        if (check1==True):
            rectangle_area(float(length), float(breadth))
            break
        
        else:
            while(True):
            
                print('INVALID INPUT. PLEASE ENTER POSITIVE NUMBERS\n')
                new_length = input('Enter the length of the rectangle: \n')
                new_breadth = input('Enter the breadth of the rectangle: \n')
                check1_repeat = input_validity(input_value1 = new_length, input_value2 = new_breadth)
                #check2_repeat = input_validity(new_breadth)
                if (check1_repeat==True):
                    rectangle_area(float(new_length), float(new_breadth))
                    break
        break       
############################################################

    elif choice=='3':
        print('Triangle area is being calculated.')
        base = input('Enter the length of base: \n')
        height = input('Enter the length of perpendicular height: \n')
        check1 = input_validity(input_value1 = base, input_value2 = height)
        #check2 = input_validity(breadth)
        if (check1==True):
            triangle_area(float(base), float(height))
            break
        
        else:
            while(True):
            
                print('INVALID INPUT. PLEASE ENTER POSITIVE NUMBERS\n')
                new_base = input('Enter the length of base: \n')
                new_height = input('Enter the length of perpendicular height: \n')
                check1_repeat = input_validity(input_value1 = new_base, input_value2 = new_height)
                #check2_repeat = input_validity(new_breadth)
                if (check1_repeat==True):
                    triangle_area(float(new_base), float(new_height))
                    break
        break
############################################################

    elif choice=='4':
        print('Circlearea is being calculated.')
        radius = input('Enter the radius of the circle: \n')
        check1 = input_validity(input_value1 = radius)
        if check1==True:
            circle_area(float(radius))
            break
        
        else:
            while(True):
                print('INVALID INPUT. PLEASE ENTER A POSITIVE NUMBER\n')
                new_radius = input('Enter the radius of the circle: \n')
                check1_repeat = input_validity(input_value1 = new_radius)
                if (check1_repeat==True):
                    cirle_area(float(new_radius))
                    break
        break       
###############################################################

   
    
        
```

### Problem 2:

Create and encrypter in python based on the ceaser cipher. It is a substitution cipher where each character of the original text is shifted a certain number characters in the alphabet. Write a function that would require 2 arguments – the input text to be encrypted and a key. For eg: Given the input text ‘hello’ and the key 3, the resulting encryted text would be ‘khoor’. Here you can see that every character in the string hello is shifted by 3 characters. ‘h’ has shifted to ‘k’, ‘e’ has shifted to ‘h’ and so on. If a key of 5 were used, the resulting string would be ‘mjqqt’. This function should be capable of ignoring any characters which are not alphabets. Th2 character ‘z’ entered b
y the user for a key of 3 would result in ‘c’.

Usage:<br>
**encrypt(‘hello world!’, 3)**<br>
**‘khoor zruog!’**

Similarly create decrypter which can decode the encryted text when provided the input text and key

Usage:<br>
**decrypt(‘khoor zruog!’, 3)**<br>
**‘hello world!’**

For the sake of simplicity you can assume that input solely consists of lowercase alphabets, spaces and punctuation symbols. Numbers in the input text would also be ignored similar to symbols.

```python
##############################################################
# ENCRYPTER_PYTHON 
##############################################################


import math

def encrypt( string_input, int_key ):
    
    result_text = ''
    print(string_input)
    #list_input = list(string_input)
    for element in string_input:
        if (ord(element) >= ord('a') and ord(element) <= ord('z')):
            
            remainder = (ord(element)+int_key) % 97
            move_by = remainder % 26
            new_ascii = 97 + move_by
            new_char = chr(new_ascii)
            result_text = result_text + new_char
            
        else:
            
            
            result_text = result_text + element
            
    
    return result_text
            

###################################################################

input_text = input('Enter the text to be encrypted: \n')   
encrypt_key = int(input('Enter the encryption key: \n'))



if ((encrypt_key > float(-math.inf)) and (encrypt_key < float(math.inf))):
        key_valid = True
        
else:
        key_valid = False
        

if key_valid == True:
    encrypted_text = encrypt(string_input=input_text.lower(), int_key=encrypt_key)
    #encrypted_list = encrypt(string_input=input_text, int_key=int(str(encrypt_key)))
    print('The result of Encryption = ', encrypted_text)
    
else:
    
    print('Invalid encryption key entered !!!')
    


        
```

```python
##############################################################
# DECRYPTER_PYTHON 
##############################################################

import math

def decrypt( string_input, int_key ):
    
    result_text = ''
    #print(string_input)
    #list_input = list(string_input)
    for element in string_input:
        if (ord(element) >= ord('a') and ord(element) <= ord('z')):
           
            remainder = (ord(element)-int_key) % 97
            
            if (ord(element)-int_key) < ord('a'):
                difference = ord('a') - (ord(element)-int_key)
                remainder = (123 - difference) % 97
                
            
            move_by = remainder % 26
            new_ascii = 97 + move_by
            new_char = chr(new_ascii)
            result_text = result_text + new_char
            
        else:
            
            result_text = result_text + element
            
    
    return result_text
            

###################################################################

input_text = input('Enter the text to be decrypted: \n')   
decrypt_key = int(input('Enter the integer decryption key: \n'))



if ((decrypt_key > float(-math.inf)) and (decrypt_key < float(math.inf))):
        key_valid = True
        
else:
        key_valid = False
        

if key_valid == True:
    decrypted_text = decrypt(string_input=input_text.lower(), int_key=decrypt_key)
    #encrypted_list = encrypt(string_input=input_text, int_key=int(str(encrypt_key)))
    print('The result of Decryption = ', decrypted_text)
    
else:
    
    print('Invalid decryption key entered !!!')
    
```
